```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sympy import *
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols
from sympy.abc import x,k,m,p
from matplotlib.ticker import FormatStrFormatter
import matplotlib
```

```python
N=1000
k=np.linspace(0,np.pi,N)
```

```python
K_2=2*np.sin(k/2)
K_4=(1/12)*(27*np.sin(k/2)-np.sin(3*k/2))
```

```python
a=[]
a.append([1/2])
a.append([2/3,1/12])
a.append([3/4,3/20,1/60])
```

```python
def general_KM(m,a,k):
    K_m=[]
    for m in a:
        sumd=0
        for t in range(1,len(m)+1):
            sumd=sumd+ ((-1)**(t+1))*m[t-1]*np.sin(t*k)
        sumd=2*sumd
        K_m.append(sumd)
    return K_m
K_m=np.array(general_KM(m,a,k))
```

```python
f,ax=plt.subplots(1)
ax.plot(k/pi,k/pi,label='k\u0394',color='black')
ax.plot(k/pi,K_m[0]/pi,label="$K_{2} \u0394$", marker='o', markersize=0.5)
ax.plot(k/pi,K_m[1]/pi,label="$K_{4} \u0394$",marker='h', markersize=0.5)
ax.plot(k/pi,K_4/pi,label="$K_{4m} \u0394$",marker='D', markersize=0.5)
ax.legend()
ax.xaxis.set_major_formatter(FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=0.25))
ax.yaxis.set_major_formatter(FormatStrFormatter('%g $\pi$'))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=0.25))
ax.set_xlabel('k \u0394')
ax.set_ylabel('K \u0394')
```

```python
K_G=(52*np.sin(k)+np.sin(2*k))/(24*np.cos(k)+30)
```

```python
f,ax=plt.subplots(1)
ax.plot(k/pi,k/pi,label='k\u0394 Analytical Limit',color='black')
ax.plot(k/pi,K_G/pi,label="$K_{(6)} \u0394$ Compact Finite Difference",color='red', marker='o', markersize=0.5)
ax.plot(k/pi,K_m[2]/pi,label="$K_{(6)} \u0394$ Central Finite Difference", marker='h', markersize=0.5)


ax.legend()
ax.xaxis.set_major_formatter(FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=0.25))
ax.yaxis.set_major_formatter(FormatStrFormatter('%g $\pi$'))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=0.25))
ax.set_xlabel("$k \u0394$")
ax.set_ylabel("$K_{(m)} \u0394$")
```
