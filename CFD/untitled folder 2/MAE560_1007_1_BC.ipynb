```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=100
x=np.linspace(-10,10,N+2)
x=x[1:-1]
dx=x[1]-x[0]
f=1+np.tanh(x)
fa=(1/np.cosh(x))**2
```

```python
plt.plot(x,fa,label='fa')
plt.plot(x,f,label='f')
plt.legend()
plt.show()
```

```python
fD_BCL = 0 
fD_BCR = 2 

fN_BCL = 0  
fN_BCR = 0 
```

```python
fn1=np.zeros(len(f))
fn1[0]=(-fD_BCL+f[1])/(2*dx)
fn1[-1] = (-f[-2] + fD_BCR ) / (2*dx)
for j in range(1,(len(f)-1)):
    fn1[j]=(-f[j-1]+f[j+1])/(2*dx)
plt.plot(x,fn1)
```

```python
BCV=np.zeros(len(f))
BCV[0]= (-fD_BCL) / (2*dx)
BCV[-1]=fD_BCR / (2*dx)
fn1=np.zeros(len(f))
fn1[0]=f[1]/(2*dx)
fn1[-1]=-f[-2]/(2*dx)
for j in range(1,(len(f)-1)):
    fn1[j]=(-f[j-1]+f[j+1])/(2*dx)
fn1 = fn1 + BCV
plt.plot(x,fn1)
```

```python
##fn2: N + D
BCV = np.zeros(len(f))
BCV[1] = + fN_BCL / (2) 
BCV[-1] = + fD_BCR / (2*dx)

fn2 = np.zeros(len(f))

fn2[0] = (-f[1] + f[2])/(2*dx)
fn2[-1] = (-f[N-1])/(2*dx)
for j in range(1,len(f)-1):
    fn2[j] = (-f[j-1] + f[j+1]) / (2*dx)

fn2 = fn2 + BCV 
plt.plot(x,0.05+fn1,label='fn1')
plt.plot(x,fn2,label='fn2')
plt.legend()
plt.show()
```

```python
##fn3: D + N
BCV = np.zeros(len(f))
BCV[0] = - fD_BCL / (2*dx) 
BCV[-1] = + fN_BCR / (2)

fn3 = np.zeros(len(f))

fn3[0] = ( f[1])/(2*dx)
fn3[-1] = (-f[-2]+f[-1])/(2*dx)
for j in range(1,len(f)-1):
    fn3[j] = (-f[j-1] + f[j+1]) / (2*dx)

fn3 = fn3 + BCV 
#plt.plot(x,fn1,label='fn1')
plt.plot(x,fn3,label='fn3')
plt.legend()
plt.show()
```
