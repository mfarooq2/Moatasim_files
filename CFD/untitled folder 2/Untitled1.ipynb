```python
import numpy as np
import matplotlib.pyplot as plt
def central_diff(y,h):
    return (y[2:]-y[:-2])/(2*h)
```

```python
# step size
N=40
x=np.linspace(0,1,N+2)
h = x[1]-x[0]

# compute function
y = np.sin(2*np.pi*x)


f_x=central_diff(y,h)
x_grid=x[1:N+1]
y_grid=y[1:N+1]
f_x_analytical=2*np.pi*np.cos(2*np.pi*x_grid)
```

```python
plt.plot(x_grid, y_grid, label = "function")
plt.scatter(x_grid, f_x, c ="green",label = "numerical_derivative")
#plt.plot(x_grid, f_x, label = "numerical_derivative")
plt.plot(x_grid, f_x_analytical, label = "analytical_derivative")
plt.legend()
plt.show()
```
